# 부분회귀
- 회귀분석 데이터에 새로운 독립변수가 추가되면 기존의 가중치값이 변한다.
- 이를 통해서 한 독립변수에서 다른 독립변수들의 영향을 제거하여 순수한 종속변수와의 상관관계를 구한다.
    - 즉 어떤 독립변수가 종속변수의 순수하게 자신만으로 어떤 영향을 미치는지 알 수 있다.

## 1. 새로운 독립변수의 추가
- x1 독립변수만으로 회귀분석 한 결과
    - $y=w_1x_1 + e$
- 새로운 독립변수 x2를 추가하여 회귀분석하면 x_1에 대한 가중치 w_1'은 w_1과 같을까?
    - $y=w_1^{'}x_1 + w_2^{'}x_2 + e^{'}$
    - 새로운 데이터가 추가 된 경우 다시 회귀분석하면 기존 독립변수에 대한 가중치는 어떻게 변할까?
- **새로운 독립변수가 추가 되면 w_1'은 w_1과 다르다.**
    - 종속변수에 영향을 미치는 모든 독립변수를 회귀모형에 포함하지 않는 한(밝혀지지 않은 독립변수들의 가능성까지) 모형의 가중치는 항상 편향된(biased) 값이다.
    
## 2. 새로운 독립변수의 추가와 가중치의 변화 증명
- 현재 데이터의 독립변수를 X1 그룹과 X2 그룹으로 나눈다.
    - X = [X1, X2]
- 독립변수 X1 만으로 회귀분석을 하는 경우의 가중치 벡터
    - $w_1 = (X_1^TX_1)^{-1} X_1^Ty$
    - OLS(최소자승법)에서 RSS(잔차제곱합)의 그레디언트 벡터가 0이라는 조건에서 나온 식
- 독립변수 X2를 추가한 새로운 선형 회귀모형
    - $y=\hat{y} + e^{'} = [X_1 \; X_2] 
    \begin{bmatrix} 
    w_1^{'} \\ 
    w_2^{'}  
    \end{bmatrix} + e^{'}$
    - w_1'과 w_2'은 독립변수 X2가 추가 된 새로운 모형의 가중치 벡터, e'은 새로운 모형의 잔차 벡터
- 직교방정식을 구하여 새로운 가중치 w_1'구할 수 있다. : 양변에 X를 곱해준다. 
    - $\begin{bmatrix} 
    X_1^{T} \\ 
    X_2^{T} 
    \end{bmatrix} [X_1 \; X_2]
    \begin{bmatrix} 
    w_1^{'} \\ 
    w_2^{'}  
    \end{bmatrix} = 
    \begin{bmatrix} 
    X_1^{T}y \\ 
    X_2^{T}y 
    \end{bmatrix}$
    - $\begin{bmatrix}
    X_1^TX_1 & X_1^TX_2 \\
    X_2^TX_1 & X_2^TX_2
    \end{bmatrix}
    \begin{bmatrix} 
    w_1^{'} \\ 
    w_2^{'}  
    \end{bmatrix} = 
    \begin{bmatrix} 
    X_1^{T}y \\ 
    X_2^{T}y 
    \end{bmatrix}$
- 직교방정식에서 w_1'을 구하면 w_1의 값과 다르다는 것을 알 수 있다.
    - $\begin{matrix} w_1^{'}
    & = (X_1^TX_1)^{-1} X_1^T (y - X_2w_2^{'}) \\
    & = (X_1^TX_1)^{-1} X_1^Ty - (X_1^TX_1)^{-1} X_1^TX_2w_2^{'} \\
    & = w_1 - (X_1^TX_1)^{-1} X_1^TX_2w_2^{'} \\
    \end{matrix}$
- **즉 새로운 독립변수가 추가된 후 회귀분석을 하면 기존 가중치의 값은 달라진다.** 
- 독립변수가 추가되도 가중치 값이 바뀌지 않는 경우 : 위의 w_1' 식에서
    - w_2' = 0 : 새로운 독립변수의 가중치가 0인 경우
    - X_1^TX_2 = 0 : 기존의 독립변수 X1과 새로운 독립변수 X_2가 직교하는 경우. 직교하면 상관관계가 없을 가능성이 높다.

## 2. 프리슈-워-로웰 정리
- `프리슈-워-로웰 정리 FWL Frisch-Waugh-Lovell` : 새로운 독립변수가 추가 되었을 경우 기존 독립변수의 가중치가 어떻게 변하는지에 대한 다른 표현
    - 1) 특정한 독립변수 그룹 X1로 종속변수 y를 선형 회귀분석하여 잔차 y*를 구한다.
    - 2) X1로 다른 독립변수 x2를 선형 회귀분석하여 나온 잔차 x2*를 구한다. 
    - 3) y*를 종속변수로, x2*을 독립변수로 선형 회귀분석해서 구한 가중치는 X1과 x2를 모두 사용하여 y를 선형 회귀분석하였을 때 x2에 대한 가중치와 같다.

- 직교방정식을 구하고 위에서 구한 w_1을 대입한다.
- 이 식에서 같은 벡터가 있는 것끼리 정리한다.
- 잔차행렬이 양 변에 있으면 잔차 행렬의 성질을 사용하여 식을 정리한다.
- 잔차행렬의 의미를 적용하여 x_2*T x_2* w = x_2*T y 로 정리된다.

## 3. 부분회귀 플롯
- `부분회귀 플롯 partial-regression-plot` : 독립변수의 갯수가 많은 경우 특정한 하나의 독립변수의 영향력을 시각화 하는 방법
    - Added Variable Plot 이라고도 함
- 부분회귀 플롯 그리는 방법 : 3번의 선형 회귀분석을 해야한다.
    - 1. 특정한 독립변수 x2를 제외한 나머지 독립변수 X1들로 종속변수 y를 선형 회귀분석하여 잔차 y*을 구한다.
    - 2. 나머지 독립변수 X1들로 특정한 독립변수 x2를 선형 회귀분석하여 잔차 x2*을 구한다.
    - 3. 잔차 x2*를 독립변수로, 잔차 y*를 종속변수로 하여 선형 회귀분석한다.
    - 이렇게 구한 x2*과 y*의 회귀분석 결과를 부분회귀 플롯이라고 한다.

## 4. 보스턴 집값 데이터로 부분회귀 플롯 분석하기

````python
from sklearn.datasets import load_boston

boston = load_boston()
dfX0 = pd.DataFrame(boston.data, columns=boston.feature_names)
dfx = sm.add_constant(dfX0)
dfy = pd.DataFrame(boston.target, columns=["MEDV"])
df = pd.concat([dfx, dfy], axis=1)

df.head()
````

````python
model_boston = sm.OLS(dfy, dfx)
result_boston = model_boston.fit()

print(result_boston.summary())
````

### 4-1. AGE 독립변수와 MEDV 종속변수의 상관관계
- 음의 상관관계가 있는 것 처럼 보인다. 

````python
%matplotlib inline

plt.figure(figsize=(10, 6))
sns.regplot(x="AGE", y="MEDV", data=df)
plt.show() ; 
````

### 4-2. statsmodels 로 부분회귀 플롯 그리기
```
plot_partregress(endog, exog_i, exog_others, data=None, obs_labels=true,
ret_coords=False)
```
- 인수
    - endog : 종속변수의 문자열
    - exog_i : 분석 대상이 되는 독립변수 문자열
    - exog_others : 나머지 독립변수 문자열의 리스트
    - data : 모든 데이터가 있는 데이터 프레임
    - obs_labels : 데이터 라벨링 여부
    - ret_coords : 잔차 데이터 반화 여부
- 부분회귀 플롯의 가로축은 어떤 독립변수에서 다른 독립변수의 영향을 제거한 일종의 "순수한 독립변수 성분"을 의미한다.
    - 즉 AGE에서 CRIM, ZN, INDUS ... 등 다른 독립변수들의 영향을 제거한 후 **순수한 AGE의 종속변수에 대한 영향과 같다.**
- 부분회귀 플롯이 아닌 단순 상관관계로 보면 음의 상관관계가 있는 것 처럼 보인다. 즉 AGE가 커질 수록 종속변수가 작아진다는 판단을 할 수 있다.
    - 그러나 부분회귀 플롯으로 AGE와 종속변수의 관계를 나타내면 아무런 상관관계가 없다는 것을 알 수 있다.

## 5. CCPR 플롯
- `CCPR component-component plus residual` : 부분회귀 플롯과 마찬가지로 하나의 변수의 영향을 살펴보기 위한 것
- 선형회귀 모형
    - $y=\hat{y} + e = w_1x_1 + w_2x_2 + \cdots + w_ix_i + \cdots + w_kx_k + e$
- CCPR은 이 성분들 중에서 xi를 가로축으로, wixi + e를 세로축으로 그린 스캐터 플롯과 같다. 
- statsmodels 패키지로 그릴 수 있다.

```
sm.graphics.plot_ccpr(result, exog_idx)
```

- 인수
    - result : 회귀분석 결과 객체 (OLS로 분석한 객체)
    - exog_idx : 분석 대상이 되는 독립변수 문자열

````python
sns.regplot(x="AGE", y="MEDV", data=df)
plt.show() ; 
````

````python
sm.graphics.plot_ccpr(result_boston, "AGE")
plt.show() ; 
````

````python
sm.graphics.plot_ccpr(result_boston, "CRIM")
plt.show() ; 
````

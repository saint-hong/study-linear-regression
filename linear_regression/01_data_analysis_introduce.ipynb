# 데이터 분석

## 데이터 분석이란
- 데이터 간의 관계를 파악한다.
- 파악된 관계로부터 원하는 새로운 데이터를 만들어 내는 과정이다.
- 데이터 분석의 종류
    - **예측** : prediction
    - **클러스터링** : clustering
    - **모사** : approximation 등
    
## 예측 prediction
- 숫자, 문서, 영상, 이미지, 음성 등의 입력데이터를 주면 데이터 분석의 결과로 다른 데이터를 출력하는 분석 방법
- **데이터 분석에서 예측의 의미는 미래의 의미를 포함하지 않는다.**
    - 시계열 분석에서는 시간상으로 미래의 데이터를 예측하는 경우 있다. 이러한 경우는 미래예측 forecasting 이라고 한다.
    
## 입력데이터와 출력데이터
- 예측 문제는 입력데이터와 출력데이터 두 가지 유형으로 분류할 수 있어야 한다.
    - 입력 데이터 : input data
    - 출력 데이터 : output data
- 입력데이터를 의미하는 용어들
    - X
    - 독립변수 independent variable
    - 특징 feature
    - 설명변수 explanatory variable
- 출력데이터를 의미하는 용어들
    - Y
    - 종속변수 dependent variable
    - 종속변수가 카테고리 값이면 라벨 label 또는 클래스 class
- 예측 성능이 입출력 데이터의 숫자와 종류에 영향을 많이 받는다. 따라서 입출력 데이터를 정확하게 파악해야 한다.
    - 어떤값의 데이터를 몇개 사용하겠다는 문제 정의가 선행되어야 한다.

## 규칙기반 방법과 학습기반 방법
- 예측문제를 풀기위한 방법론
- `규칙기반 rule-based`
    - 입력 데이터에 대해서 어떤 출력 데이터가 나오도록 규칙이나 알고리즘을 사람이 미리 만들어 놓는 방법
- `학습기반 training-based or 데이터 기반 data-based`
    - 대량의 입력 데이터를 주면 컴퓨터가 스스로 규칙을 만들게 하는 방법
    
## 지도학습
- 학습기반 예측 방법론의 학습용 데이터 집합(training data set)을 사람이 직접 만들어 주는 방법
- 학습용 데이터에 출력 데이터(목표값)를 붙이는 작업을 사람이 수동으로 해주어야 한다.
    - 개 사진 : "개" 목표값
    - 고양이 사진 : "고양이" 목표값
- 지도학습의 비유    
    - 학습용 데이터 : 정답이 표시된 수백 개의 문제집
    - 지도학습 : 문제집을 컴퓨터에 주고 학습 시킨다.
    - 학습 : 컴퓨터가 문제집의 문제를 나름의 방법으로 풀고 정답을 얼마나 맞혔는지 채점한다.
    - 평가 : 풀이방법으로 바꾸어 보면서 풀이와 채점을 반복한다.
- `목표값 target` : 학습용 데이터에 붙어있는 출력 데이터, 정답, 라벨
- `레이블링 labelling` : 입력데이터만 있고 목표값이 없는 경우, 각 입력데이터에 사람이 원하는 목표값을 붙여주는 작업

## 전처리와 인코딩
- `전처리 preprocessing or 인코딩 encoding` : 숫자, 문서, 이미지, 음성, 영상 등의 여러가지 데이터를 컴퓨터가 처리할 수 있는 숫자 데이터로 바꾸어주는 과정
    - **컴퓨터가 직접 처리할 수 있는 데이터는 사실상 숫자이다.**
    - 또는 여러개의 입력 데이터를 조합하여 새로운 입력데이터를 만드는 작업
    - 또는 여러개의 입력 데이터 중에서 출력데이터에 영향을 미칠 만한 핵심 데이터를 선택하는 작업 
- **전처리 과정은 최종 예측 성능에 커다란 영향을 끼친다.**

## 인코딩의 다른 예 : 문서 데이터
- 문서 데이터는 각 데이터별로 크기가 다르다. 이것을 같은 크기의 숫자로 바꾸어 주어야 한다.
    - 이미지는 nxm 개의 픽셀값으로 이루어져 있다.
    - 문서는 길이가 다 제각각이다.
- `BOW Bag of Words` : 단어의 순서, 의미 등의 정보는 무시하고 특정한 단어가 문서에 몇 번 나오는지 세고 그 빈도를 벡터로 표시하는 방법, 일반적으로 사용되는 방법이다.
    - 10000개의 단어에 1부터 10000의 번호를 붙이고, 각 문서에서 **몇 번째 단어가 몇 번이 사용되었는지를** 숫자로 나타낸다.
    - A 문서를 bow 벡터로 나타내면 : [1, 2, 0, 4] : 첫 번쨰 단어 1번, 두번째 단어 2번, 3번째 단어 0번, 4번째 단어 4번 사용했다는 의미
    - 이것을 숫자 벡터로 바꾸어 준다. 

#### 문서 데이터 임포트

```python
from sklearn.datasets import fetch_20newsgroups
```

```python
news = fetch_20newsgroups()

print("입력 : \n", news.data[0])
print("출력 : \n", news.target_names[news.target[0]])
```

#### CountVectorizer 
- 문서 데이터를 숫자로 변환해주는 기능이 있다.
- 사이킷런의 feature_extraction 의 text 모듈에 CountVectorizer 가 들어 있다. 

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vec = CountVectorizer(stop_words="english").fit(news.data[:100])
vec
```

#### 100개의 문서마다 6288개의 값이 들어 있다.
- CountVectorizer가 100개의 문서에서 등장하는 단어를 6288개 찾았다고 볼 수 있다.
- 각 문자가 몇 번 사용되었는지를 나타낸다. BOW

```python
data = vec.transform(news.data[:100])
data.shape
```

#### 일부 데이터를 시각화하면
- 50개의 문서와 100개의 단어만 사용
- 각 행은 문서, 각 열은 단어
    - 색이 진할 수록 해당 단어가 많이 사용 됐다는 것을 의미한다.

```python
plt.figure(figsize=(12, 8))

sns.heatmap(data.toarray()[:50, :100], cmap=mpl.cm.bone_r,
           linewidths=0.002, linecolor="gray", cbar=False,
           xticklabels=False, yticklabels=False)
plt.xlabel("단어")
plt.ylabel("문서")
plt.title("BOW : CountVectorizer : 인코딩 결과 시각화")
plt.show() ; 
```

## 카테고리값
- 컴퓨터가 다룰 수 있는 데이터 : 숫자, 카테고리값(=범주형 값)
- `카테고리값 category` : 숫자가 아닌 주로 기호로 표시되고 비연속적인 데이터이다.
- 숫자와 카테고리값의 차이 : 크기나 가치 혹은 순서를 비교할 수 있는가?
    - 10cm vs 20cm : 크기 비교 가능 : 숫자 데이터
    - "개" vs "고양이" : 크기 비교 불가 : 카테고리 데이터
- `클래스 class` : 카테고리값이 가질 수 있는 경우의 수
    - 동전 던지기 : 앞면, 뒷면 : **이진 클래스 binary class**
    - 주사위 던지기 : 1~6 : 세 개이상의 경우의 수 : **다중 클래스 multi class**
- 카게고리값이어도 숫자 처럼 비교가 가능한 경우도 있다. 성적, 등급, 평점 등과 같은 경우 기호로 표시 되지만 크기나 순서를 비교할 수 있다. 이러한 경우는 숫자로 바꾸어서 표시하기도 하지만 일반적인 카테고리 값으로 표시하기도 한다.

## 회귀분석과 분류
- 예측 문제는 출력하고자 하는 데이터가 숫자값인지 카테고리값인지에 따라서 사용하는 방법이 완전히 달라진다.
- `회귀분석 regression analysis` : 출력하고자 하는 값이 숫자값인 경우
    - 정답값을 직접 구하는 문제를 푸는 것과 같다.
- `분류 classification` : 출력하고자 하는 값이 카테고리값인 경우
    - 4지 선다형 객관식 문제를 푸는 것과 같다.
    - 대부분의 머신러닝 방법이 분류 문제를 풀기 위한 방법이다.

## 회귀분석의 예
- boston 집값 데이터를 사용한 회귀분석의 예

#### 주택가격, 방의갯수, 노후화, 범죄율의 상관관계 분포

```python
boston = load_boston()
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df["주택가격"] = boston.target

g = sns.pairplot(df[["주택가격", "RM", "AGE", "CRIM"]])
g.fig.suptitle("보스턴 주택가격 데이터 일부 (RM:방개수, AGE:노후화, CRIM:범죄율)", y=1.02)
plt.show() ; 
```

```python
df.head()
```

## SVC Support Vector Machine
- 서포트벡터머신 분류 모형으로 분류

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

#### 꽃잎의 길이, 꽃잎의 폭 feature만 사용
- df에서 가져온 것이 아니라 iris 원본 데이터에서 직접 가져옴
- X 입력 데이터
- y 라벨 데이터, 정답

```python
features = [2, 3]
X = iris.data[:, features]
y = iris.target
```

```python
X[:10]
```

```python
y[:10]
```

#### svc 모델 생성 및 학습

```python
model = SVC(kernel="linear", random_state=0)
model.fit(X, y)
```

#### 예측
- 꽃잎의 길이와 폭의 데이터에서 각각 최소값, 최대값을 범위로 하는 데이터를 생성하여 입력 데이터로 사용함. 

```python
XX_min = X[:, 0].min() - 1
XX_max = X[:, 0].max() + 1
YY_min = X[:, 1].min() - 1
YY_max = X[:, 1].max() + 1

XX, YY = np.meshgrid(np.linspace(XX_min, XX_max, 1000),
                    np.linspace(YY_min, YY_max, 1000))
ZZ = model.predict(np.c_[XX.ravel(), YY.ravel()]).reshape(XX.shape)
ZZ
```

```python
ZZ[0]
```

- XX.ravel() : XX를 하나의 열벡터로 만든 값
- np.c_[a, b] : a와 b를 concatnate 해서 열로 이어 붙인 값

## 비지도 학습
- `비지도 학습 unsupervised learning` : 입/출력 구분이 되지 않는 단순한 데이터관계에서 특정한 규칙을 찾아내는 방법
    - **데이터에 따라서 입력과 출력의 관계가 명확하지 않을 수 있다.**
    - 지도학습 supervised learning : 입력값과 출력값의 쌍(pair)을 학습데이터로 하여 입력값에 대한 출력값을 예측하도록 학습(training)을 시킨다.
- 입력이나 출력 데이터를 구분짓지 않고, 데이터가 입력되면 데이터들간의 규칙을 찾거나 미리 지정한 규칙에 맞는 데이터인자를 **구분해 낸다.**

## 클러스터링
- `클러스터링 clustering` : 비지도학습의 대표적인 방법, 데이터들을 유사한 데이터까지 같은 그룹으로 모으는 방법
    - **affinity propagation**

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import AffinityPropagation
```

#### 특징 벡터 2개, 클러스터 3개

```python
X, _ = make_blobs(n_features=2, centers=3, random_state=1)
X
```

```python
_
```

```python
len(X), len(_)
```

#### AffinityPropagation 클러스터링 모델 생성

```python
model = AffinityPropagation().fit(X)
model
```

```python
model.cluster_centers_indices_[0]
```

```python
model.cluster_centers_indices_[1]
```

```python
model.cluster_centers_indices_[2]
```

#### 클러스터링 결과 확인

```python
%matplotlib inline
```

```python
plt.figure(figsize=(8, 6))
plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1], marker="o", s=10, edgecolor="k")
plt.title("before clustering")

plt.subplot(122)
plt.scatter(X[:, 0], X[:, 1], marker="o", s=10, edgecolor="k")
plt.title("after clustering")

for k in range(3) : 
    # 클러스터의 센터의 위치 : [a, b]
    cluster_center = X[model.cluster_centers_indices_[k]]
    
    # 센터에서 각 점까지를 선으로 연결
    # center의 x, y 값과 데이터의 x,y값을 연결
    for x in X[model.labels_ == k] : 
        plt.plot([cluster_center[0], x[0]], [cluster_center[1], x[1]], c="k")
        
plt.suptitle("Affinity Propagation 방법을 사용한 클러스터링 결과", y=1.03)
plt.tight_layout()
plt.show() ; 
```

# 서포트 벡터 머신 : SVM

## SVM
** 정의 **
- SVM : Support Vector Machine
   - 판별함수 모형의 하나이다.
   - 퍼셉트론은 판별경계가 유니크하지 않은 단점이 있다.
   - 서포트 벡터 머신은 퍼셉트론 기반의 모형에 안정적인 판별함수(decison hyperplane)를 찾기위해 제한 조건을 추가한 모형
- 판별함수 f(x) = w^Tx - w_0 = 0
   - 벡터 w 위의 임의의 점(cW)을 지나고 벡터 w와 직교하는 어떤 직선 A의 방정식과 같다.
   - w_0 는 방정식에서 c||w||^2을 변환한 임의의 값
- 서포트 벡터란?
   - 클래스별 데이터 중에서 판별함수(판별경계)와 가장 가까운 데이터
   - x+, x-
   - 각 클래스 y=1, y=-1의 데이터 샘플중 판별함수와 가까운 것
- 마진이란?
   - 판별함수로부터 두 서포트 벡터와의 거리
   - (x+와 판별함수의 거리 - x-와 판별함수의 거리)
   - 마진= 2 / ||w||
   - 마진값이 클 수록 경계선이 더 안정적이다. 
- SVM의 목적함수는 마진값을 크게 하는 ||w||의 최소값을 찾는 방정식
   - 2차 곡선 : L= 1/2||w||^2 = 1/2 (w^Tw)
- 목적함수에 제한조건이 추가 된다.
   - 모든 데이터 x1~xN의 판별함수 값이 1보다 크거나, -1 보다 작아야 한다.
   - y_i f(x_i) = y_i (w^Tx_2 - w_0) >= 1
   - y_i 는 +1 또는 -1이다.
- 목적함수에 제한조건을 추가하기 위해 라그랑주 승수법을 사용한다.
   - L = 1/2(w^Tw) - sum a_i{y_i (w^Tx_i - w_0) -1}
   - a_i는 f(x_i)에 대한 라그랑주 승수값
   - 이 최적화 문제를 풀어서 w_0, w, a를 구하고 판별함수에 대입하면 최종적인 SVM의 판별함수를 알 수 있다. 
- 라그랑주 승수법으로 만든 제한조건이 있는 목적함수 L을 최적화하기 위해서 KKT 조건을 적용한다.
   - 부등식 제한 조건에 의해서 서포트 벡터에 해당하지 않는 모든 데이터의 라그랑지 승수 a_i는 0 값이 된다.
   - a_i = 0 (if x_i not support vector)
- 최적화의 조건
   - 목적함수 L을 w와 w_0로 미분한 값이 0이어야 한다.
      - L = 1/2(w^Tw) - sum a_i{y_i (w^Tx_i - w_0) -1}
   - L을 w와 w_0로 각각 미분하면 두 식을 구할 수 있다. 
   - 두 식을 L에 대입하여 정리하면 w와 w0가 사라지고, a_i, a_j, x_i, x_j, y_i, y_j로 이루어진 식이 된다.
   - a_i를 구하는 식으로 변환된다. = 듀얼형식
   - 즉 제한조건이 추가된 목적함수 L을 최적화하기 위해 각 변수로 미분하면 목적함수가 듀얼형식이 된다. a_i를 구하는 식으로 간편해진다.
- 듀얼 형식을 풀어 L을 최소화하는 a를 구하면, 판별함수는 다음과 같아진다.
   - f(x) = w^Tx - w_0 = sum a_iy_ix_i^Tx - w_0
   - 여기에서 y_i는 +1 또는 -1 이므로 
   - f(x) = a^+ x^T x^+ - a^- x^T x^- - w_0 이된다.
   - 즉  x와 x+ 서포트와의 코사인 유사도, x와 x- 서포트와의 코사인 유사도의 측정에 따라서 큰 쪽으로 클래스를 판별하게 된다.
- 즉 SVM 모형은 어떤 x에 대해, 두 서포트 벡터와의 코사인 유사도를 계산하여 더 이 값이 더 큰 쪽으로 클래스를 결정한다.

** SVM의 판별함수의 수학적 정의 요약 **
- 판별함수(판별경계선) f(x)는 어떤 직선 A의 방정식
- 서포트 일떄와 서포트가 아닐떄의 판별함수 값에 대한 부등식 정의
- 판별함수와 각각의 서포트와의 거리의 합을 마진이라고 할때, 마진이 클 수록 좋은 판별함수이다.
   - 직선과 점의 거리 공식을 사용하여 합한다.
   - margin : 2 / ||w||
- 최대의 margin을 얻기위해서 ||w||를 최소화하는 목적함수를 도입한다.
- 목적함수 L은 가중치 벡터 w의 최소값을 얻기위한 함수이다.
- 서포트 벡터가 아닌 나머지 데이터 샘플에 대한 제한조건을 목적함수에 추가한다.
- 라그랑지 승수법으로 제한조건이 추가된 목적함수 L을 최적화 한다는 것은 변수 w, w_0로 미분한 값이 0인 것을 찾는 것과 같다.
   - 라그랑지 승수 a_i는 서포트 벡터가 아닌 나머지 데이터인 경우 값이 0이 된다는 조건이 추가된다.
- 목적함수 L을 w, w_0로 각각 편미분 하면 두개의 식을 얻을 수 있다.
- 두 개의 식을 제한조건이 추가 된 목적함수 L에 대입하여 정리하면 a_i, y_i, x_i, w_0로 이루어진 듀얼형식이 된다.
   - a_i를 구하면 최적화를 할 수 있게 된다.
- 듀얼형식으로 간편해진 목적함수 L을 최소화하는 a_i를 사용하여 판별함수 f(x)를 다시 정리한다.
   - f(x) = sum a_i y_i x^T x_i - w_0
- 이 식에서 라그랑지 승수의 조건을 사용하면, 서포트 벡터가 아닌 x_i의 경우 a_i = 0이므로 판별함수가 의미가 없어진다.
- 따라서 판별함수 f(x)는 x_i가 서포트 벡터인 경우에만 남게된다.
   - f(x) = (a^+ y^+ x^T x^) + (a^- y^- x^T x^-) - w_0
   - y는 +1 과 -1 값이므로 대입하면 
   - f(x) = (a^+ x^T x^) - (a^- x^T x^-) - w_0 이된다.
- 최종 판별함수 식에서, x^T와 각 서포트 벡터와의 코사인 유사도의 크기에 따라서 클래스가 판별된다는 것을 알 수 있다.

** SVM 슬랙변수 **
- 슬랙변수 : 
   - 데이터가 직선이 판별경계선으로 나누어지지 않는다면? 선형분리로 불가능한 샘플의 경우 슬랙변수를 사용하여 개별 오차를 허용하여 분류를 쉽게 만들어준다.
   - 즉 완전히 선형분리가 안되어 클래스를 예측하기 애매한 경우의 샘플들.
   - 슬랙변수 엡실론은 0보다 크거나 같다.
- 일반 샘플 x_i의 판별함수값 정의에 슬랙변수를 적용한다.
- 이러한 슬랙변수를 적용한 최종 목적함수 L (제한조건이 추가된)
   - L = 1/2 ||w||^2 - sum a_i(y_i (w^Tx_i - w_0) -1 + 엡실론_i) - sum mu_i엡실론_i + C sum 엡실론_i
   - C sum 엡실론_i는 슬랙변수의 합이 커지지 않도록 제어해준다.
- SVC 모형의 파라미터 C는 슬랙변수의 크기를 제어하는 상수이다.
   - 이 값에 따라서 서포트 벡터의 갯수가 달라진다. 즉 판별하기 어려운 데이터를 서포트로 묶어 준다.

** SVC support vector classifier **
- from sklearn.svm import SVC
- SVC의 파라미터
   - kernel : 커널 선택
   - C : 슬랙변수 가중치 선택
- SVC의 속성값
   - n_support_ : 각 클래스별 서포트 벡터의 갯수
   - support_ : 각 클래스별 서포트 벡터의 인덱스
   - support_vectors_ : 각 클래스별 서포트 벡터의 x값
   - coef_ : 가중치 w 벡터
   - intercept_ : -w0
   - dual_coef_ : 각 원소가 ai, yi로 이루어진 벡터 (목적함수의 듀얼형식으로의 변환)
- 판별함수값 = 클래스 예측값 = 듀얼 형식으로 구한 값
   - 판별함수값이 0보다 작으면 calss 1, 0보다 크면 class 2
   - svc.decision_function([x_new])
   - svc.coef_.dot(x_new) + svc.intercept_ : w^Tx - w_0
   - 듀얼형식 (a_i, y_i로 이루어진 판별함수)
      - svc.dual_coef[0][0] * svc.support_vectors_[0].dot(x_new) + \
        svc.dual_coef[0][1] * svc.support_vectors_[1].dot(x_new) + \
        svc.intercept_
- 슬랙변수 가중치 C 값의 작아질 수록 서포트 벡터의 갯수가 늘어난다.

## kernel SVM

** 커널 서포트 벡터 머신 **
- 퍼셉트론과 SVM은 XOR 문제를 풀 수 없다.
   - 선형판별평면(decision hyperplane)이므로 데이터를 분류할 수 없다.
- 이러한 문제를 해결하기 위해서 데이터 x를 기저함수 벡터로 변환한다.
   - D차원 벡터인 x를 M차원 기저함수 벡터로 변환
   - x_1 = (1, 2) 일때, 기저함수 벡터 phi(x) = (phi_1(x), phi_2(x), phi_3(x)) 로 변환하는 것.
   - 2차원 벡터가 3차원 벡터가 된다.
- 기저함수 phi(x)를 목적함수 L과 판별함수 f(x)에 각각 대입한다.
   - 즉 x_i, x_j에 phi(x_i), phi(x_j)를 대입한다.
- 이렇게 대입하면 phi(x_i)^T phi(x_j)의 내적으로 묶이게 된다
   - kernel = k(x_i, x_j) = phi(x_i)^T phi(x_j)
- 즉 커널함수는 기저함수의 내적으로 이루어진 함수이다.
- 목적함수 L과 예측모형(판별함수)에 커널함수를 대입한다.
   - k(x_i, x_j)
- 커널함수를 사용하지 않으면 원래 식대로 x_i^Tx_j 가 된다. 
- 커널함수의 특징 : k(x_i, x_j)
   - x_i, x_j가 같은 벡터이면 최대가 된다.
   - x_i, x_j의 거리가 멀 수록 작아진다.
   - x_i, x_j의 코사인 유사도를 측정하는 것과 같다. 
- 커널의 장점
   - 베이시스 함수 계산량이 줄어든다.
- 커널의 확장과 생성
   - 커널함수를 사용하여 다양한 커널함수들을 정의할 수 있다. 만들 수 있다. 
   - 커널 * C (C>0) = 커널
   - 커널 + C (C>0) = 커널
   - 커널 * 커널 = 커널
   - 커널 + 커널 = 커널
   - 시그모이드(커널) = 커널 (단조증가하는 함수 = 시그모이드 함수)
   - x1, x2 각각의 커널의 곱 = 커널
- 많이 사용되는 커널
   - 선형 서포트 벡터 머신 : k(x1, x2) = x1^Tx2 (두 벡터의 내적)
   - 다항 커널 (polynomial kernel) : k(x1, x2) = (gamma(x1^Tx2) + theta)^d
      - 벡터의 내적으로 정의된 커널을 확장한 것
   - RBF 커널 (Radial Basis Function) : k(x1, x2) = exp(-gamma||x1 - x2||^2) ---> 가우시안 커널
   - 시그모이드 커널 (sigmoid kernel) : k(x1, x2) = tahn(gamma(x1^Tx2) + theta)
- 대표적인 커널들은 각각 기저함수로 분리된다.

** 커널 SVM 요약 **
- SVM의 선형판별평면을 다변화하기 위하여 (유연하게) 기저함수를 도입한 것
   - D차원의 데이터 샘플이 kernel 함수를 통과하면 M차원의 벡터가 된다.
- SVM의 목적함수 L과 판별함수(예측모형) f(x)에서 xi^Txj 를 커널함수로 대치한다.
   - k(xi, xj)
- 커널의 장점
   - 기저함수의 계산 시간이 빠르다.
- 커널함수의 확장
   - 커널함수는 양수 C, 커널간의 연산을 통해서 새로운 커널을 만들 수 있다.
- 대표적인 커널함수들
   - 선형서포트벡터머신 : 두 벡터의 내적
   - 다항커널 : 내적에 gamma를 곱한 후 theta를 더한 값에 d 자승을 한 함수 (여러가지 차수의 그래프가 기저함수)
   - RBF 커널 : x1과 x2 사이의 거리에 gamma를 곱한 값을 지수함수에 넣은 함수 
   - 시그모이드 커널 : 내적에 gamma를 곱하고 theta를 더한 값에 than 함수를 적용한 함수

** kernel 설정 **
- SVC 클래스에서 kernel 파라미터를 지정하여 kernel 함수를 설정할 수 있다.
- kernel
   - linear : 선형 SVM
   - poly : 다항 커널
      - gamma : gamma 값
      - coef(theta) : theta 값
      - degree : 차수
   - rbf or None : RBF 커널
      - gamma : gamma 값
   - sigmoid : 시그모이드 커널
      - gamma : gamma 값
      - coef(theta) : theta 값

   - kernel="sigmoid" SVC가 성능이 대체로 떨어진다.

## sklearn docs SVM
- 분류, 회귀, 이상치 감지에 사용 된다.
   - classificatino
   - regression
   - outliers detection
- SVM의 장점
   - 고차원 데이터 샘플 공간에서 효과적이다.
   - 차원 수가 표본수(데이터 샘플)보다 많은 경우에도 효과적이다. (features > samples)
   - 판별함수(결정함수, decision function)에서 훈련 데이터의 서포트 벡터를 사용하므로 메모리를 효율적으로 사용
   - 다양한 커널을 지정할 수 있다. : 사용자 정의 커널도 지원 (custom kernel)
- SVM의 단점
   - 차원수가 표본수보다 훨씬 많은 경우 커널 함수를 선택할때 과적합을 피하기 위해 정규화 항이 중요하다.(C?)
   - 확률추정치값(probability estimators)을 제공하지 않는다.
- SVM 입력값
   - dense(밀도) 샘플과 sparse(희소) 샘플 벡터를 모두 사용가능.
   - 희소행렬을 입력값으로 사용할 때는 dtype=float64, C-ordered 또는 scipy.sparse.csr_matrix(spars)를 사용해야함
- 종류
   - SVC, NuSVC, LinearSVC, SVR, NuSVR, LinearSVR 모형이 있다.
   - kernel에 따라서 여러 종류의 SVC 가능 
      - linear, poly, rbf, sigmoid
   - 이진, 다중 클래스 분류를 수행할 수 있다.
- LinearSVC
   - SVC, NuSVC와 달리 속성값이 다르거나 없다. : support_
   - 손실 함수로 squared_hinge 손실을 사용한다
   - intercept_scaling 파라미터를 조정해서 절편의 정규화 효과를 줄일 수 있다(?)
- SVM 특징
   - 최대 마진으로 초평면(yperplane, 판별경계면, 판별경계선)을 판단한다.
- multi-class 문제
   - decision_function_shape 파라미터의 값 설정 : {"ovo", "ovr"}
   - multi-class 인 경우 one-vs-one 방식으로 계산한다. : 모든 클래스의 2가지 조합을 계산하는 방식
   - one-vs-rest 방식으로 설정할 수 있다. : 표본이 해당 클래스에 포함하는지 안하는지 여부를 계산하는 방식(이진클래스가 된다.)
   - clf.decision_function([[1]]) 의 shape 속성값으로 ovr, ovo의 계산 횟수를 확인 할 수 있다. 
   - LinearSVC의 다중 클래스 문제
      - multi_class="crammer_singer" 로 설정하여 다중 클래스 문제를 풀 수 있으나, 일반적으로 ovr 방법이 사용됨.
- Scores and Probabilities
   - 모형의 성능값과 클래별 예측 확률값의 크기 인덱스가 다를 수 있다. 
   - 매서드 predict_proba 보다 decision_function() 매서드를 사용하는 것이 좋을 수 있다. 
- 비대칭 데이터 (Unbalanced problems)
   - class_weight(클래스에 가중치 부여) 또는 sample_weight(샘플 데이터에 가중치 부여) 파라미터를 설정하여 특정 클래스의 샘플에 더 많은 중요성을 부여할 수 있다. 
   - 비대칭 데이터의 경우 이와같이 class_weight, sample_weight 를 설정하여 결정경계를 더 포괄적으로 만들 수 있다. 
- Tips and Practical Use
   - 입력 데이터의 복사(copy) 피하기 : SVC, SVR, NuSVC, NuSVR은 입력 데이터를 복사하게끔 되어 있다.
      - liblinear의 희소데이터로 사용된다(?)
      - 대규모 데이터의 선형분류인 경우 SGDClassifier를 사용하는 것이 효과적이다. 목적함수는 LiearSVC와 거의 동일하다.
   - 커널의 캐시 크기 : 대규모 데이터 분류시 커널의 캐시의 크기가 커진다. 이러한 경우 RAM 사용을 기본값 200MB 보다 더 높게 설정하는 것이 좋다. 500MB~1000MB
   - C 설정(슬랙변수의 크기제어 상수) : 기본 1이 가장 적합하다. 노이즈가 많은 데이터인 경우 값을 줄인다. C의 크기가 줄어들면 더 많은 정규화에 해당한다. 즉 fitting에 노이즈 데이터도 사용된다.
      - C의 크기가 커지면 훈련시간이 길어진다.
   - 데이터의 스케일링 : SVM 알고리즘은 scale 불변 이 아니므로 대규모 데이터 셋은 scale을 추천한다. 입력 데이터 X의 모든 값을 [0, 1] 또는 [1, -1]로 변환한다. 아니면 StandardScaler를 사용하여 평균=0, 분산=1 로 만든다. 테스트 벡터에도 동일하게 적용한다. 
      - clf = make_pipeline(StandardScaler(), SVC())
   - shrinkage parameter : 반복횟수가 많은 경우 shrinkage 값은 훈련시간을 단축시켜줄 수 있다. 최적화 문제를 느슨하게 적용하면(예를들어 stopping parameter 값을 크게 해두면) shrinkage 값을 적용하지 않아도 훈련이 빠를 수 있다.
   - 데이터의 불균형 : 데이터가 불균형한 경우 많은 positive와 적은 negative인 경우, class_weight="balanced"를 사용하면 효과적이다. 또는 C parameter를 사용하여 조정.
- 다양한 커널 함수들
   - linear
   - polynomial
   - rbf : RBF 커널을 더 확장성 있게 사용할 수 있다.
   - sigmoid
- SVM의 중요한 파라미터 C, gamma
   - RBF 커널의 사용시 C와 gamma 파라미터를 설정해야한다.  
   - C는 모든 SVM 모델에 공통적으로 적용되는 매개변수이며, 의사결정경계의 단순성, 훈련예제의 잘못된 분류를 상쇄시켜준다.
   - C값이 작으면 : 의사결정표면을 매끄럽게 만든다.
   - C값이 크면 : 모든 데이터 샘플을 정확하게 분류하려고 한다.
   - gamma는 개별 데이터 하나가 훈련에 미치는 영향을 조절한다. 
   - gamma가 클 수록 다른 데이터 샘플에 영향을 많이 받는다(?) 
- 커스텀 커널을 사용할 수 있다.
   - kernel 값으로 내가 만든 커널 함수를 적용할 수 있다.  
   - kernel="precomputed" : 그람행렬처럼 미리 계산된( X * X.T) 함수를 커널로 적용할 수 있다. 
- 기타 특징
   - LinearSVC는 선형판별모형으로 dual 형식이 아니므로 linear kernel 외에 다른 커널을 적용할 수 없다.
   - NuSVC는 C-SVC의 변형 : nu 파라미터는 서포트 벡터의 수와 마진 에러의 수를 제어한다. C와 유사한 역할.
      - nu 값은 마진 오차의 분수에 대한 상한, 서포트 벡터의 분수에 대한 하한 값 (?)
- RBF SVM의 C와 gamma 특징
   - gamma
      - 개별 샘풀의 영향력의 정도를 나타낸다. gamma 값이 낮으면 멀고, 높으면 가깝다. support vector의 영향력 반경의 역수와 같다.
      - 모형은 gamma의 값에 민감하다.
      - gamma 값이 크면 support vector의 영향은 support 벡터에만 해당된다.
      - gamma가 매우 작아지면 모형이 제약을 많이 받게되어 데이터의 복잡성을 포착하지 못할 수 있다. support vectore의 영향에 모든 샘플이 포함된다.
   - C
      - 마진을 최대화하는 것에 대한 샘플 분류의 효과를 상쇄한다. C값이 클수록 모든 샘플들을 올바르게 분류하도록 하고, C값이 작을수록 마진이 더 작아진다. C값이 작을 수록 정확도가 낮아지지만 더 큰 마진이 적용된다. 
      - C는 과적합을 방지할 수 없다. 
      - C가 크면 훈련시간이 길어질 수 있다. 
      - C가 작으면 support vector가 많아져 예측 시간이 증가될 수 있다. (support vector와의 코사인 유사도 계산량이 많아지므로)
      - C값이 작으면 피팅 시간은 줄지만, 예측시간이 높아지게 된다. (트레이드 오프)
   - gamma가 중간정도이고, C가 매우 커지면 성능이 높아 질 수 있다. 
   - gamma와 C는 반비례 관계에서 대체로 좋은 성능이 나타난다. 
   - gamma가 중간값이고 C가 큰 경우도 좋은 성능이 나타날 수 있다.

### XOR 데이터 생성

```python
np.random.seed(0)
```

```python
X_xor = np.random.randn(200, 2)
X_xor[:10]
```

```python
# X 중에서 0보다 큰 값을 골라내기

y_xor = np.logical_xor(X_xor[:, 0] > 0, X_xor[:, 1] > 0)
y_xor
```

```python
# T이면 1, F이면 0으로 변환

y_xor = np.where(y_xor, 1, 0)
y_xor
```

```python
from sklearn.svm import SVC
```

```python
def plot_xor(X, y, model, title, xmin=-3, xmax=3, ymin=-3, ymax=3) : 
    
    XX, YY = np.meshgrid(np.arange(xmin, xmax, (xmax-xmin)/1000),
                        np.arange(ymin, ymax, (ymax-ymin)/1000))
    ZZ = np.reshape(model.predict(
        np.array([XX.ravel(), YY.ravel()]).T), XX.shape)
    
    plt.contourf(XX, YY, ZZ, cmap=mpl.cm.Paired_r, alpha=0.5)
    plt.scatter(X[y==1, 0], X[y==1, 1], c="b", marker="o", label="class=1", s=50)
    plt.scatter(X[y==0, 0], X[y==0, 1], c="r", marker="s", label="class=0", s=50)
    plt.xlim(xmin, xmax)
    plt.ylim(ymin, ymax)
    plt.title(title)
    plt.xlabel("x1")
    plt.ylabel("x2")
```

```python
%matplotlib inline
```

```python
polysvc = SVC(kernel="poly", degree=2, gamma=1, coef0=0).fit(X_xor, y_xor)
rbfsvc = SVC(kernel="rbf").fit(X_xor, y_xor)
sigmoidsvc = SVC(kernel="sigmoid", gamma=2, coef0=2).fit(X_xor, y_xor)

plt.figure(figsize=(8, 12))
plt.subplot(311)
plot_xor(X_xor, y_xor, polysvc, "다항커널 SVC를 사용한 분류 결과")
plt.subplot(312)
plot_xor(X_xor, y_xor, rbfsvc, "RBF 커널 SVC를 사용한 분류 결과")
plt.subplot(313)
plot_xor(X_xor, y_xor, sigmoidsvc, "Sigmoid 커널 SVC를 사용한 분류 결과")

plt.tight_layout()
plt.show() ;
```

## SVM-Anova : feature selection

- iris 데이터 샘플에 36개의 feature를 추가했을 때 몇 개의 feature를 선택하면 분류 성능이 가장 높을지 판단할 수 있다. 
   - form sklearn.feature_selectiono import SelectPercentile

```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)

X.shape, y.shape
```

### 이상치 추가
- X : 150x4 의 행렬
- sample_rng : 150x36 행렬
- np.hstack((X, sample_rng)) : 150x40 : 4+36 

```python
rng = np.random.RandomState(0)
sample_rng = 2 * rng.random((X.shape[0], 36))
X = np.hstack((X, sample_rng))

X.shape
```

```python
X[0], sample_rng[0]
```

### pipeline

```python
from sklearn.feature_selection import SelectPercentile, f_classif
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
clf = Pipeline(
    [
        ("anova", SelectPercentile(f_classif)),
        ("scaler", StandardScaler()),
        ("svc", SVC(gamma="auto"))
    ]
)
clf
```

### SelectPercentile 값에 따른 fitting

```python
from sklearn.model_selection import cross_val_score
```

```python
score_means = list()
score_stds = list()
percentiles = (1, 3, 6, 10, 15, 20, 30, 40, 60, 80, 100)

for percentile in percentiles : 
    clf.set_params(anova__percentile=percentile)
    scores = cross_val_score(clf, X, y)
    score_means.append(scores.mean())
    score_stds.append(scores.std())
```

```python
score_means
```

```python
score_stds
```

```python
plt.errorbar(percentiles, score_means, np.array(score_stds))
plt.hlines(np.max(score_means), 
           xmin=0, xmax=100, linestyle="--", linewidth=0.7, color="r")
plt.vlines(percentiles[np.argmax(score_means)], 
           ymin=0.75, ymax=1.0, linestyle="--", linewidth=0.7, color="r")
plt.xlabel("percentiles")
plt.ylabel("accuracy score")
plt.title(
    "performence of the SVM-Anova varying the percetiles of \n"
    "features selected"
)
plt.grid(False)
plt.xlim(-1, 101)
plt.ylim(0.75, 1.0)
plt.show() ; 
```

## SVM with Custom kernel
- kernel 함수를 내가 만들고 적용할 수 있다. 

```python
from sklearn.datasets import load_iris
from sklearn.inspection import DecisionBoundaryDisplay
```

```python
iris = load_iris()
X = iris.data[:, :2]
y = iris.target

X.shape, y.shape
```

```python
def my_kernel(X, Y) : 
    
    M = np.array([[2, 0], [0, 1]])
    return np.dot(np.dot(X, M), Y.T)
```

```python
np.dot(test_1, y[:2].T)
```

### 불균형 데이터에 대한 데이터 샘플별 가중치 적용
- 개별 데이터 샘플에 대한 가중치값을 array로 만든 후 fit의 매개변수 sample_weight를 설정한다.
- 샘플 가중치가 같은 경우와 샘플 가중치가 다른 경우 SVM의 판별경계가 달라진다. 

```python
def plot_decision_function(clf, sample_weight, axis, title) : 
    
    xx, yy = np.meshgrid(np.linspace(-4, 5, 500), 
                        np.linspace(-4, 5, 500))
    Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    axis.contourf(xx, yy, Z, alpha=0.75, cmap=plt.cm.bone)
    axis.scatter(X[:, 0], X[:, 1], c=y, s=100 * sample_weight, alpha=0.9,
                cmap=plt.cm.bone, edgecolor="k")
    axis.axis("off")
    axis.set_title(title)
```

```python
np.random.seed(0)
X = np.r_[np.random.randn(10, 2) + [1, 1], 
          np.random.randn(10, 2)]
X[:10]
```

```python
X.shape
```

```python
y = [1] * 10 + [-1] * 10
y
```

```python
sample_weight_last_ten = abs(np.random.randn(len(X)))
sample_weight_last_ten
```

```python
sample_weight_constant = np.ones(len(X))
sample_weight_constant
```

```python
sample_weight_last_ten[15:] *= 5
```

```python
sample_weight_last_ten[9] *= 15
```

```python
sample_weight_last_ten
```

```python
clf = SVC(gamma=1)
clf.fit(X, y)
```

```python
swclf = SVC(gamma=1)
swclf.fit(X, y, sample_weight=sample_weight_last_ten)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
plot_decision_function(clf, sample_weight_constant, axes[0], "constant weight")
plot_decision_function(swclf, sample_weight_last_ten, axes[1], "modified weight")
plt.tight_layout()
plt.show() ; 
```
